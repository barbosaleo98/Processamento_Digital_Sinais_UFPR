```python
# Inicializa os pacotes que serão utilizados
import numpy as np
import matplotlib.pyplot as plt
```

1) fazer uma função que calcule a convolução entre dois sinais (vetores). (sem usar a função conv() ).

```python
# Cria uma função que realiza a convolução.
# Determina-se o comprimento dos vetores x e h que são
# armazenados como M e N respectivamente.
# Estes então determinam a quantidade de elementos 
# do vetor de saída, que terá M+N-1 zeros inicialmente. 
# Os vetores X e H são x e h concatenados com um vetor de
# zeros, para que tenham o mesmo tamanho sempre.

# Em seguida são utilizados dois loops for.
# O primeiro corresponde ao índice do vetor y
# e varre toda a extensão deste.

# O segundo for é responsável por realizar as operações 
# de convolução para cada elemento de y. O k corresponde ao 
# índice da somatória da convolução discreta.

def convolucao(x,h):
    N = len(x)
    M = len(h)  
    y=np.zeros(M+N-1)
    X = np.concatenate((x,np.zeros(M)),axis=None)
    H = np.concatenate((h,np.zeros(N)),axis=None)
   
    for n in range(N+M-1):
        for k in range(M):
            if(0<=n-k):
                y[n] = y[n] + X[k]*H[n-k] 
    return y
```

2) testar com as seguintes entradas, plotando com a sua função e a função conv( ): 

2.a ) x=[1 0 1], h=[3 2 1]

```python
x = [1, 0, 1]
h = [3, 2, 1]

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Comparação entre as saídas das funções de convolução", fontsize=16)
axarr[0].stem(convolucao(x,h),basefmt = " ",use_line_collection=True)
axarr[0].set_title('Função criada')
axarr[0].grid()
axarr[1].stem(np.convolve(x,h),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Função do pacote Numpy')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```
