```python
# Inicializa os pacotes que ser√£o utilizados
import numpy as np
import matplotlib.pyplot as plt
```

a) Construa o vetor a, considerando um sinal senoidal de 100Hz amostrado a uma taxa de 1000 A/s durante 0,2 s.


```python
fsig = 100
fsample = 1000
T = 0.2
t = np.linspace(0,T,fsample*T)
a = np.sin(2*np.pi*fsig*t)

# Imprime o gr√°fico
plt.grid(True)
plt.plot(t,a)
plt.title('Senoide')
plt.show()
```

b) Construa o vetor b, considerando um sinal composto pela soma de tr√™s senoides: 60, 110, 160 Hz, com amplitudes 3, 2 e 1, amostrado a uma taxa de 1000 A/s durante 0,2 s.


```python
fsig1 = 60
fsig2 = 110
fsig3 = 160
fsample = 1000
T = 0.2
tb = np.linspace(0,T,fsample*T)
b1 = np.sin(2*np.pi*fsig1*tb)
b2 = np.sin(2*np.pi*fsig2*tb)
b3 = np.sin(2*np.pi*fsig3*tb)
b = b1 + b2 + b3

# Imprime o gr√°fico
plt.grid(True)
plt.plot(tb,b)
plt.title('Senoide')
plt.show()
```

c) Construa o vetor c a partir da decima√ß√£o do vetor b, jogando fora uma amostra para cada amostra. Calcule o vetor C, como a T.F. do vetor c.


```python
N = int(fsample*T/2) #Quantidade de amostras
tc = np.linspace(0,T,N)
c = np.zeros(N)
j=0

for i in range(2*N):
    if i%2 == 0:
        c[j] = b[i]
        j += 1
        
C = np.fft.fft(c)

#Imprime o gr√°fico
plt.grid(True)
plt.plot(tc,c)
plt.title('Senoide')
plt.show()

#Imprime o gr√°fico
plt.grid(True)
plt.plot(abs(C))
plt.title('Senoide')
plt.show()
```

d) Construa o vetor d a partir da decima√ß√£o do vetor c, jogando fora uma amostra para cada amostra. Calcule o vetor D, como a T.F. do vetor d.


```python
N = int(fsample*T/4) #Quantidade de amostras
td = np.linspace(0,T,N)
d = np.zeros(N)
j=0

for i in range(2*N):
    if i%2 == 0:
        d[j] = c[i]
        j += 1
        
D = np.fft.fft(d)

#Imprime o gr√°fico
plt.grid(True)
plt.plot(td,d)
plt.title('Senoide')
plt.show()
#Imprime o gr√°fico
plt.grid(True)
plt.plot(abs(D))
plt.title('Senoide')
plt.show()
```

e) plote numa figura 3x2 (com subplot(3,2,...) os sinais (b), (c), (d): na primeira coluna o sinal no tempo e na segunda coluna o sinal em frequ√™ncia (valor absoluto).

```python
B = np.fft.fft(b)

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(3, 2)
fig.suptitle("Compara√ß√£o entre os vetores b,c e d", fontsize=16)
axarr[0,0].set_title('b no tempo')
axarr[0,0].plot(tb,b)
axarr[1,0].plot(tc,c)
axarr[1,0].set_title('c no tempo')
axarr[2,0].plot(td,d)
axarr[2,0].set_title('d no tempo')
axarr[0,1].plot(abs(B))
axarr[0,1].set_title('B na frequ√™ncia')
axarr[1,1].plot(abs(C))
axarr[1,1].set_title('C na frequ√™ncia')
axarr[2,1].plot(abs(D))
axarr[2,1].set_title('D na frequ√™ncia')
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

f) construa o vetor f, inserindo zeros entre as amostras do vetor c. Calcule F=fft(f)

```python
N = int(fsample*T) #Quantidade de amostras
tf = np.linspace(0,T,N)
f = np.zeros(N)
i=0

for j in range(N):
    if j%2 == 0:
        f[j] = c[i]
        i += 1
    else:
        f[j] =0

F = np.fft.fft(f)


#Imprime o gr√°fico
plt.grid(True)
plt.plot(tf,f)
plt.title('Senoide')
plt.show()

#Imprime o gr√°fico
plt.grid(True)
plt.plot(abs(F))
plt.title('Senoide')
plt.show()
```
