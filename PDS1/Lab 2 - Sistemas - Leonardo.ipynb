```python
# Inicializa os pacotes que ser√£o utilizados
import numpy as np
import matplotlib.pyplot as plt
```

1) Implemente os seguintes sistemas: </br>

1.a ) atrasa a entrada em duas unidades

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o n√∫mero de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que ir√° ser preenchido pela fun√ß√£o sistA
out = np.array(np.zeros(N))

# Fun√ß√£o sistA que desloca os valores de sa√≠da 2 unidades para a direita 
# Para i<=1 n√£o √© poss√≠vel deslocar 2 unidades, pois n√£o existe valor de
# x anterior a x[1]. Assim um if √© necess√°rio
def sistA(x,N):  
    for i in range (N):
        if i <= 1 :
            out[i] = 0
        else:
            out[i] = x[i-2]
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Atraso da entrada em duas unidades", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistA(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Sa√≠da')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.b ) M√©dia-m√≥vel: Cada valor da sa√≠da seja a m√©dia aritm√©tica do valor de entrada e suas 2 posi√ß√µes anteriores.

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o n√∫mero de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que ir√° ser preenchido pela fun√ß√£o sistB
out = np.array(np.zeros(N))

# Fun√ß√£o sistB que calcula a m√©dia aritm√©tica entre o valor 
# atual de x e 2 posi√ß√µes anteriores. Em i=1 e i=2 o c√°lculo 
# da m√©dia √© realizado com a quantidade de valores dispon√≠veis, 
# sendo que # a partir de i=3 sempre haver√° pelo menos 2 
# valores anteriores dispon√≠veis at√© i=N. 

def sistB(x,N):
    for i in range (N):
        if i == 1 :
            out[i] = x[i]/3
        elif i==2:
            out[i] = (x[i]+x[i-1])/3
        else:
            out[i] = (x[i]+x[i-1]+x[i-2])/3
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("M√©dia m√≥vel", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistB(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Sa√≠da')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.c ) a sa√≠da corresponde ao valor quadr√°tico da entrada

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o n√∫mero de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que ir√° ser preenchido pela fun√ß√£o sistB
out = np.array(np.zeros(N))

# Fun√ß√£o sistC cuja sa√≠da √© a entrada elevada ao quadrado usando a fun√ß√£o
# power() do pacote Numpy.
def sistC(x,N):
    for i in range (N):
        out[i] = np.power(x[i],2)
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Sa√≠da quadr√°tica", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistC(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Sa√≠da')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.d ) a sa√≠da corresponde a uma combina√ß√£o linear da entrada multiplicada por 2 mais a entrada deslocada de 1 unidade multiplicada por 3

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o n√∫mero de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que ir√° ser preenchido pela fun√ß√£o sistB
out = np.array(np.zeros(N))

# Fun√ß√£o sistD que realiza a combina√ß√£o linear entre os vetores modificados de x.
# Para i=1 n√£o √© poss√≠vel deslocar, pois n√£o existe valor de x anterior a x[1]. 
# Assim um if √© necess√°rio
def sistD(x,N):
    for i in range (N):
        if i==1:
            out[i] = (x[i]*2)
        else:
            out[i] = (x[i]*2) + (x[i-1]*3)
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Combina√ß√£o linear", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistD(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Sa√≠da')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.e ) a sa√≠da corresponde ao dobro da entrada somada ao valor 5

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o n√∫mero de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que ir√° ser preenchido pela fun√ß√£o sistB
out = np.array(np.zeros(N))

# Fun√ß√£o sistE que dobra o valor de x e soma 5
def sistE(x,N):
    for i in range (N):
        out[i] = (x[i]*2)+5
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Dobro da entrada somada ao valor 5", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistE(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Sa√≠da')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

 1.f ) Acumulador. A sa√≠da correponde a acumula√ß√£o (soma) dos valores da entrada at√© o presente momento.

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o n√∫mero de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que ir√° ser preenchido pela fun√ß√£o sistB
out = np.array(np.zeros(N))

# Fun√ß√£o sistF que soma a sa√≠da os valores da entrada anteriores. 
# Note que como n√£o h√° valores anteriores a x[1], em i=1 a sa√≠da out[1] √© 
# composta somente por x[1].
def sistF(x,N):
    for i in range (N):
        if i == 1:
            out[i] = x[i]
        else:
            out[i] = out[i-1] + x[i]
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gr√°ficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Acumulador", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistF(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Sa√≠da')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```
