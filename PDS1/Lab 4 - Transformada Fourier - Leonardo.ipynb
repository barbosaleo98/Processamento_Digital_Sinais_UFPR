```python
# Inicializa os pacotes que serão utilizados
import numpy as np
import matplotlib.pyplot as plt
```

1) Implementar uma função X=tfourier(x,N) que calcule a N pontos da transformada de fourier (de zero a 2pi)  da sequencia de números dada pelo vetor x. Ou seja, o vetor X vai conter N elementos, cada um correspondente a um valor de w (0 a 2pi).

```python
# Cria uma função tfourier(x,N) que recebe como argumentos um vetor 
# x e uma quantidade de pontos N. Dentro da função, o vetor x 
# tem seu comprimento ajustado para ter N pontos independente do
# tamanho original dele. Isso é feito ao concatenar zeros ao final
# de x até que este tenha comprimento N.

# Também é criado um vetor y, de comprimento igual ao x ajustado.
# y inicialmente é preenchido com zeros, para depois receber os 
# valores novos.

# Os loops "for" são utilizados para aplicar a definição da 
# transformada de Fourier. O loop de índice n corresponde 
# aos pontos no domínio do tempo (vetor x), enquanto o k 
# corresponde ao pontos no domínio da frequência (vetor y).

# Para cada valor de k de 0 a N-1, é realizada a somatória
# dos valores x[n] com n de 0 a N-1 vezes a exponencial 
# complexa responsável pela troca de base.

def tfourier(x,N):
    x = np.concatenate((x, np.zeros(N - len(x))), axis=None)
    y = np.zeros(len(x))
    for k in range(N):
        y[k] = 0
        for n in range(N):
            y[k] = y[k] + x[n]*np.exp((-1j*2*np.pi*(k-1)*(n-1))/N)
    return y
```

2) Comparar (plotar) o resultado (valor absoluto) com a função X=fft(x) para as  seguintes entradas:

ao plotar os dois graficos com diferentes numero de pontos, utilize o mesmo eixo normalizado (entre 0 e 1). Plote
a curva com menor numero de pontos apenas com simbolo 'o', e a curva com mais pontos com linha continua.

ex. plot((0:(N-1))/N, abs(tfourier(x,N)), (0:length(X)-1)/length(X),abs(X),'o')

2.a) N=100; x=[zeros(1,10) ones(1,10) zeros(1,10)];

```python
# Cria o vetor x e o número N de pontos. Em seguida realiza 
# as transformadas de Fourier com a função criada (y1)
# e com a função do pacote numpy (y2).
x = np.concatenate((np.zeros(10),np.ones(10),np.zeros(10)), axis=None)
N = 100
y1 = tfourier(x,N)
y2 = np.fft.fft(x)

# Cria os vetores horizontais de comprimento N que serão 
# posteriormente normalizados no comando plot
l1 = np.linspace(0,N-1,N)
l2 = np.linspace(0,len(y2)-1,len(y2))

# Plota o gráfico comparativo. y1 é plotado como linha
# contínua e y2 somente com 'o's.
plt.grid(True)
plt.plot(l1/N,abs(y1))
plt.plot(l2/len(y2),abs(y2),'o')
plt.legend(['Função criada','Função do pacote Numpy'])
plt.title("Comparativo entre a transformada criada e a do Numpy")
plt.show()
```

2.b) N=200; n=0:49; x=sin(pi/8*n);

```python
# Cria o vetor x e o número N de pontos. Em seguida realiza 
# as transformadas de Fourier com a função criada (y1)
# e com a função do pacote numpy (y2).
N = 200
n = np.linspace(0,48,49)
x = np.sin(np.pi/8*n)
y1 = tfourier(x,N)
y2 = np.fft.fft(x)

# Cria os vetores horizontais de comprimento N que serão 
# posteriormente normalizados no comando plot
l1 = np.linspace(0,N-1,N)
l2 = np.linspace(0,len(y2)-1,len(y2))

# Plota o gráfico comparativo. y1 é plotado como linha
# contínua e y2 somente com 'o's.
plt.grid(True)
plt.plot(l1/N,abs(y1))
plt.plot(l2/len(y2),abs(y2),'o')
plt.legend(['Função criada','Função do pacote Numpy'])
plt.title("Comparativo entre a transformada criada e a do Numpy")
plt.show()
```
