```python
# Processamento Digital de Sinais I - TE352
# LAB 1 - Sinais
# Leonardo Barbosa da Silva GRR20162855

# 1) Gerar os seguintes sinais e visualiza-los (comandos plot e stem). 
# Gerar 100 valores.
# Inicializa os pacotes que serão utilizados
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 1.a) Impulso unitário

# Cria um vetor y composto de zeros
y=np.zeros(100)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(-50, 49, 100)

# Atribui o valor 1 ao elemento central do vetor y, 
# representando um impulso unitário
y[50]=1

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Impulso unitário')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Impulso unitário')
plt.show()
```

```python
# 1.b) Degrau unitário
# Cria um vetor de 50 zeros 
y0=np.zeros(50)

# Cria um vetor de 50 elementos, todos de valor um
y1=np.ones(50)

# Concatena os vetores y0 e y1, de forma que o vetor 
# de zeros vem primeiro
y=np.append(y0,y1)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(-50, 50, 100)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Degrau unitário')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Degrau unitário')
plt.show()
```

```python
# 1.d) Senoide com 4 ciclos

# Cria um vetor de 100 pontos igualmente espaçados, 
# com valor final 
# correspondente a 2*pi*n, sendo n o valor de ciclos
x=np.linspace(0, 2*np.pi*4, 100)

# Função seno em função do vetor x
y=np.sin(x)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Senoide com 4 ciclos')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Senoide com 4 ciclos')
plt.show()
```

```python
# 1.e) Cosenoide com 5 ciclos

# Cria um vetor de 100 pontos igualmente espaçados, 
# com valor final 
# correspondente a 2*pi*n, sendo n o valor de ciclos
x=np.linspace(0, 2*np.pi*5, 100)

# Função cosseno em função do vetor x
y=np.cos(x)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Cosenoide com 5 ciclos')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Cosenoide com 5 ciclos')
plt.show()
```

```python
# 1.f) Exponencial complexa com 4 ciclos

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(0, 100, 100)

# Define a frequência como 4/100, para que em 100 pontos
# a função exponencial complete 4 ciclos
f=4/100

# Transforma a frequência de Hz para rad/s
w=2*np.pi*f

# Função exponencial complexa
y=np.exp(1j*w*x)

# Cria ambientes de figura, utilizando subplots para 
# imprimir dois gráficos em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Exponencial complexa com 4 ciclos", fontsize=16)
axarr[0].plot(x, np.real(y))
axarr[0].set_title('Parte real')
axarr[1].plot(x, np.imag(y),'orange')
axarr[1].set_title('Parte imaginária')
fig.tight_layout()
fig.subplots_adjust(top=0.88)
plt.show()                 

# Semelhante ao bloco acima, porém com dados discretos ao 
# longo do tempo.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Exponencial complexa com 4 ciclos", fontsize=16)
axarr[0].stem(x, np.real(y))
axarr[0].set_title('Parte real')
axarr[1].stem(x, np.imag(y),'orange')
axarr[1].set_title('Parte imaginária')
fig.tight_layout()
fig.subplots_adjust(top=0.88)
plt.show()  
```

```python
# 1.h) Senoide de 60 Hz (amostragem = 300 Hz)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(1, 100, 100)

# Converte a freq. de amostragem no tempo entre as amostras 
Ts=1/300

# Frequência angular do sinal senoidal
w=2*np.pi*60

# Sinal senoidal em função de x, w e tempo entre amostras
y=np.sin(w*Ts*x)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Senoide de 60 Hz (amostragem = 300 Hz)')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Senoide de 60 Hz (amostragem = 300 Hz)')
plt.show()
```

```python
# 1.i) Onda quadrada com 4 ciclos

# Cria um vetor de 100 pontos igualmente espaçados
x = np.linspace(1, 100, 100)

# Define a frequência como 4/100, para que em 100 pontos
# o sinal quadrado complete 4 ciclos
f= 4/100

# Gera a onda quadrada em função de x e de f
y = signal.square(2 * np.pi * f * x)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Onda quadrada com 4 ciclos')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Onda quadrada com 4 ciclos')
plt.show()
```

```python
# 1.j) Onda triangular com 3 ciclos

# Cria um vetor de 100 pontos igualmente espaçados
x = np.linspace(0,100,100)

# Define a frequência como 4/100, para que em 100 pontos
# o sinal quadrado complete 4 ciclos
f = 3/100

# Define a frequência angular w e o sinal dente de serra
# com largura 0.5 (onda simétrica)
w = 2*np.pi*f
y = signal.sawtooth(w*x, 0.5)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y)
plt.title('Onda triangular com 3 ciclos')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.title('Onda triangular com 3 ciclos')
plt.show()
```

```python
# 2) Operações
# 2.a) Deslocar o impulso unitario (a) duas unidades para a direita. 
# Plotar impulso original e impulso deslocado em cores diferentes.

# Cria vetores de 100 zeros cada
y=np.zeros(100)
y_deslocado = np.zeros(100)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(-50, 49, 100)

# Atribui o valor 1 ao elemento central do vetor y, 
# representando um impulso unitário
y[50] = 1

# Atribui o valor correspondente ao impulso do vetor y no 
# vetor y_deslocado, porém dois elementos a direita
y_deslocado[50+2] = y[50]

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y,x,y_deslocado)
plt.title('Impulso unitário e impulso deslocado 2 unidades para a direita')
plt.show()

plt.grid(True)
plt.stem(x, y)
plt.stem(x, y_deslocado,'orange')
plt.title('Impulso unitário e impulso deslocado 2 unidades para a direita')
plt.show()
```

```python
# 2.b) Somar impulso unitario e 3 vezes o impulso
# unitario deslocado a direita.

# Cria vetores de 100 zeros cada
y=np.zeros(100)
y_deslocado = np.zeros(100)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(-50, 49, 100)

# Atribui o valor 1 ao elemento central do vetor y, 
# representando um impulso unitário
y[50] = 1

# Atribui o valor correspondente ao impulso do vetor y no 
# vetor y_deslocado, porém dois elementos a direita
y_deslocado[50+2] = y[50]

# Soma o impulso unitario e 3 vezes o impulso unitario
# deslocado a direita.
y_soma=y+(3*y_deslocado)

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y_soma)
plt.title('Somar impulso unitario e 3 vezes o impulso unitario deslocado a direita')
plt.show()

plt.grid(True)
plt.stem(x, y_soma)
plt.title('Somar impulso unitario e 3 vezes o impulso unitario deslocado a direita')
plt.show()
```

```python
# 2.c) Construir uma rampa somando impulsos unitarios
# deslocados de 0; 1; 2 e 3 unidades.

# Cria vetores de 100 zeros cada
y=np.zeros(100)
y_deslocado_1 = np.zeros(100)
y_deslocado_2 = np.zeros(100)
y_deslocado_3 = np.zeros(100)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(-50, 49, 100)

# Atribui o valor 1 ao elemento central do 
# vetor y, representando um impulso unitário.
y[50] = 1

# Realiza somas dos impulsos deslocados com 
# os impulsos anteriores
y_deslocado_1[50+1] = y[50] + y[50]
y_deslocado_2[50+2] = y[50] + y_deslocado_1[50+1]
y_deslocado_3[50+3] = y[50] + y_deslocado_2[50+2]

# Soma os sinais resultantes
y_soma=y+y_deslocado_1+y_deslocado_2+y_deslocado_3

# Imprime os gráficos "contínuos" e discretos
plt.grid(True)
plt.plot(x,y_soma)
plt.title('Rampa de impulsos deslocados')
plt.show()

plt.grid(True)
plt.stem(x, y_soma)
plt.title('Rampa de impulsos deslocados')
plt.show()
```

```python
# 2.d)Subtrair um degrau unitário deslocado à direita 
# de uma unidade de um degrau unitário.

# Cria um vetor de 50 zeros
y0=np.zeros(50)


# Cria um vetor de 50 elementos, todos de valor um
y1=np.ones(50)

# Concatena os vetores y0 e y1, de forma que o vetor
# de zeros vem primeiro
y=np.append(y0,y1)

# Cria um vetor de 100 pontos igualmente espaçados
x=np.linspace(-51, 50, 100)

# Loop que desloca o degrau uma unidade para a direita
for i in range(1,100):
    y_deslocado[i] = y[i-1]

# Subtrai o degrau unitário deslocado à direita de uma
# unidade de um degrau unitário.
y_subtracao=y-y_deslocado

# Imprime os gráficos "contínuos" e discretos.
plt.grid(True)
plt.plot(x,y_subtracao)
plt.title('Somar impulso unitario e 3 vezes o impulso unitario deslocado a direita')
plt.show()

plt.grid(True)
plt.stem(x, y_subtracao)
plt.title('Somar impulso unitario e 3 vezes o impulso unitario deslocado a direita')
plt.show()
```
