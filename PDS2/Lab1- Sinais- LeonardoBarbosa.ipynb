```python
# Inicializa os pacotes que serão utilizados
import numpy as np
import matplotlib.pyplot as plt
```

<b>1) Gerar os seguintes sinais e visualizá-los (comandos plot e stem). Gerar 200 valores para n (de 0 a 199).</b>

```python
# Vetor n  com 200 valores igualmente espaçados
n = np.linspace(0,199,200)
```

<b>a)</b> senoide com frequência w=0,2 rad ( sin(w*n) )

```python
# Criação de variáveis para a 
# frequência angular e sinal senoidal
w = 0.2
x = np.sin(n*w)

# Imprime os gráficos 
plt.grid(True)
plt.plot(n,x)
plt.title('Senoide com frequência w=0,2 rad - plot')
plt.show()

plt.grid(True)
plt.stem(n, x,use_line_collection=True)
plt.title('Senoide com frequência w=0,2 rad - stem')
plt.show()
```

<b>b)</b> senoide com 8 ciclos

```python
# Frequência do sinal em Hz que resultará em 
# 8 ciclos dentre os 200 pontos de n
f = 8/200

# Frequência angular com base na frequência f
w = 2*np.pi*f

# Crição do sinal senoidal
x = np.sin(n*w)

# Imprime os gráficos
plt.grid(True)
plt.plot(n,x)
plt.title('Senoide com 8 ciclos - plot')
plt.show()

plt.grid(True)
plt.stem(n, x,use_line_collection=True)
plt.title('Senoide com 8 ciclos - stem')
plt.show()
```

<b>c)</b> cosenoide com 3 ciclos

```python
# Frequência do sinal em Hz que resultará em 
# 3 ciclos dentre os 200 pontos de n
f = 3/200

# Frequência angular com base na frequência f
w = 2*np.pi*f

# Crição do sinal cossenoidal
x = np.cos(n*w)

# Imprime os gráficos 
plt.grid(True)
plt.plot(n,x)
plt.title('Cossenoide com 3 ciclos - plot')
plt.show()

plt.grid(True)
plt.stem(n, x,use_line_collection=True)
plt.title('Cossenoide com 3 ciclos - stem')
plt.show()
```

<b>d)</b> senoide de 60 Hz (frequência de amostragem = 800 amostas por segundo = 800 Hz)

```python
# Converte a freq. de amostragem no tempo entre as amostras 
Ts=1/800

# Frequência angular do sinal senoidal
w=2*np.pi*60

# Sinal senoidal em função de n, w e tempo entre amostras
x=np.sin(w*Ts*n)

# Imprime os gráficos 
plt.grid(True)
plt.plot(n,x)
plt.title('Senoide de 60 Hz (amostragem = 800 Hz) - plot')
plt.show()

plt.grid(True)
plt.stem(n, x, use_line_collection = True)
plt.title('Senoide de 60 Hz (amostragem = 800 Hz) - stem')
plt.show()
```

<b>e)</b> exponencial complexa y=exp(j*w*n), calcular w para 3 ciclos. Plotar parte real e a parte imaginária

```python
# Define a frequência como 3/200, para que em 200 pontos
# a função exponencial complete 3 ciclos
f=3/200

# Transforma a frequência de Hz para rad/s
w=2*np.pi*f

# Função exponencial complexa
y=np.exp(1j*w*n)

# Cria ambientes de figura, utilizando subplots para 
# imprimir dois gráficos em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Exponencial complexa com 3 ciclos - plot", fontsize=16)
axarr[0].plot(n, np.real(y))
axarr[0].set_title('Parte real')
axarr[1].plot(n, np.imag(y),'orange')
axarr[1].set_title('Parte imaginária')
fig.tight_layout()
fig.subplots_adjust(top=0.88)
plt.show()                 

# Semelhante ao bloco acima, porém com dados discretos ao 
# longo do tempo.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Exponencial complexa com 3 ciclos - stem", fontsize=16)
axarr[0].stem(n, np.real(y), use_line_collection = True)
axarr[0].set_title('Parte real')
axarr[1].stem(n, np.imag(y),'orange', use_line_collection = True)
axarr[1].set_title('Parte imaginária')
fig.tight_layout()
fig.subplots_adjust(top=0.88)
plt.show()  
```

<b>f)</b> exponencial oscilante crescente. y=alfa.^n (ou y=alfa**n; y=pow(alfa,n)). Escolher valor complexo de alfa. Plotar parte real

```python
# Alfa definido arbitrariamente, para que seu módulo seja maior que 1
alfa = 0.28 +1j
mod_alfa = round(abs(alfa),2)

# Gera o sinal exponencial oscilante crescente
y=alfa**n

# Imprime os gráficos 
plt.grid(True) 
plt.plot(n, np.real(y))
plt.title('Exponencial oscilante crescente (parte real) - plot')
plt.legend(['y com |\u03B1|: ' + str(mod_alfa)])
plt.show()

plt.grid(True)
plt.stem(n, np.real(y), use_line_collection = True)
plt.title('Exponencial oscilante crescente (parte real) - stem')
plt.legend(['y com |\u03B1|: ' + str(mod_alfa)])
plt.show()
```
