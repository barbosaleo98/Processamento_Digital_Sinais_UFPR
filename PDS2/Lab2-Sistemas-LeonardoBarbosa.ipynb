<b>1) Crie função para implementar os seguintes sistemas que produzem um sinal de saída a partir de um sinal de entrada:</b>

<b>a)</b> Atrasador: atrasa a entrada em duas amostras

```python
# Cria o vetor x que será deslocado
x = array([0, 0, 0, 1, 2, 3, 2, 1, 0, 0, 0])

# Função shift_right que desloca os valores de saída em n unidades para a direita 
def shift_right(x,n): 
    if not isinstance(n, int):
        print('n deve ser um numero inteiro')
        return x
    out = array(zeros(len(x)))
    for i in range (len(x)):
        if i-n > len(x)-1:
            break
        else:
            out[i] = x[i-n]
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = subplots(2, 1)
fig.suptitle("Atraso da entrada em duas unidades", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(shift_right(x,2),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()
```

<b>b)</b> Média-móvel: Cada valor da saída seja a média aritmética do valor de entrada e suas 2 posições anteriores.

```python
# Cria o vetor x que será utilizado para cálculo da média móvel
x = array([0, 0, 0, 1, 2, 3, 4, 0, 0, 0, 0])

# Função moving_average que calcula a média-móvel 
def moving_average(x, n):
    if n < 1 or not isinstance(n, int):
        print('n deve ser um numero inteiro <= 1')
        return x
    res = cumsum(x, dtype=float)
    res[n:] = res[n:] - res[:-n]
    return res[n - 1:] / n  
    
# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = subplots(2, 1)
fig.suptitle("Média móvel de janela = 3", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(moving_average(x,3),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()
```

<b>c)</b> Quadrador: a saída corresponde ao valor quadrático da entrada

```python
# Cria o vetor x que será transformado
x = array([0, 0, 0, 1, 2, 3, 2, 1, 0, 0, 0])

# Função quad que retorna o valor quadrático da entrada
def quad(x): 
    out = array(zeros(len(x)))
    for i in range (len(x)):
        out[i] = x[i]**2
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = subplots(2, 1)
fig.suptitle("Quadrador", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(quad(x),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()
```

<b>d)</b> Combinador: a saída corresponde a uma combinação linear da entrada multiplicada por 2 mais três vezes a entrada atrasada de 1 amostra.

```python
# Cria o vetor x que será transformado
x = array([0, 0, 0, 1, 2, 3, 0, 0, 0, 0, 0])

# Função combinator que realiza a combinação linear entre os vetores modificados de x.
# Para i=1 não é possível deslocar, pois não existe valor de x anterior a x[1]. 
# Assim um if é necessário
def combinator(x):
    out = array(zeros(len(x)))
    for i in range (len(x)):
        if i==1:
            out[i] = (x[i]*2)
        else:
            out[i] = (x[i]*2) + (x[i-1]*3)
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = subplots(2, 1)
fig.suptitle("Combinador", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(combinator(x),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()
```

<b>e)</b> Dobrador com offset: a saída corresponde ao dobro da entrada somada ao valor 5.

```python
# Cria o vetor x que será transformado
x = array([0, 0, -1, 1, 2, 3, 2, 1, -1, 0, 0])

# Função doubler que dobra a entrada e soma 5
def doubler(x): 
    out = array(zeros(len(x)))
    for i in range (len(x)):
        out[i] = (x[i]*2) + 5
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = subplots(2, 1)
fig.suptitle("Dobrador com offset de 5 unidades", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(doubler(x),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()
```

<b>f)</b> Acumulador: A saída corresponde a acumulação (soma) dos valores do sinal de entrada até o presente momento.

```python
# Cria o vetor x que será transformado
x = array([0, 0, 0, 1, 2, 3, 2, 1, 0, 0, 0])

# Função accumulator que realiza soma cumulativa 
def accumulator(x): 
    out = array(zeros(len(x)))
    for i in range (len(x)):
        if i == 1:
            out[i] = x[i]
        else:
            out[i] = out[i-1] + x[i]
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = subplots(2, 1)
fig.suptitle("Acumulador", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(accumulator(x),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()
```

<b>2)</b> Considere os sinais com 200 amostras:

w=4*2$\pi$/200

x1=sin(w*n)

x2=sign(x)

x3=x1+x2

Crie uma função que plota no gráfico da esquerda os sinais x1,x2,x3 e no gráfico da direita os sinais y1+y2 e y3,

onde y1=T{x1}, y2=T{x2}, y3=T{x3}, e T{} é cada um dos sistemas do item anterior.

```python
n = linspace(0,199,200)
w = 4*2*pi/200
x1 = sin(w*n)
x2 = sign(x1)
x3 = x1 + x2

# y do Atrasador
y1 = shift_right(x1,2)
y2 = shift_right(x2,2)
y3 = shift_right(x3,2)

# Plots do Atrasador
fig, axarr = subplots(1, 2)
fig.suptitle("Atrasador", fontsize=16)
axarr[0].plot(n,x1,n,x2,n,x3)
axarr[0].set_title('Entradas')
axarr[0].legend(['x1','x2','x3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[0].grid()
axarr[1].plot(n,y1+y2,'--',n,y3)
axarr[1].set_title('Saídas')
axarr[1].legend(['y1+y2','y3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()

# y da Média Móvel
y1 = moving_average(x1,3)
y2 = moving_average(x2,3)
y3 = moving_average(x3,3)

# Plots da Média Móvel
fig, axarr = subplots(1, 2)
fig.suptitle("Média Móvel de Janela 3", fontsize=16)
axarr[0].plot(n,x1,n,x2,n,x3)
axarr[0].set_title('Entradas')
axarr[0].legend(['x1','x2','x3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[0].grid()
axarr[1].plot(n[:-2],y1+y2,'--',n[:-2],y3)
axarr[1].set_title('Saídas')
axarr[1].legend(['y1+y2','y3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()

# y do Quadrador
y1 = quad(x1)
y2 = quad(x2)
y3 = quad(x3)

# Plots do Quadrador
fig, axarr = subplots(1, 2)
fig.suptitle("Quadrador", fontsize=16)
axarr[0].plot(n,x1,n,x2,n,x3)
axarr[0].set_title('Entradas')
axarr[0].legend(['x1','x2','x3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[0].grid()
axarr[1].plot(n,y1+y2,'--',n,y3)
axarr[1].set_title('Saídas')
axarr[1].legend(['y1+y2','y3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()

# y do Combinador
y1 = combinator(x1)
y2 = combinator(x2)
y3 = combinator(x3)

# Plots do Combinador
fig, axarr = subplots(1, 2)
fig.suptitle("Combinador", fontsize=16)
axarr[0].plot(n,x1,n,x2,n,x3)
axarr[0].set_title('Entradas')
axarr[0].legend(['x1','x2','x3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[0].grid()
axarr[1].plot(n,y1+y2,'--',n,y3)
axarr[1].set_title('Saídas')
axarr[1].legend(['y1+y2','y3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()

# y do Dobrador
y1 = doubler(x1)
y2 = doubler(x2)
y3 = doubler(x3)

# Plots do Dobrador
fig, axarr = subplots(1, 2)
fig.suptitle("Dobrador", fontsize=16)
axarr[0].plot(n,x1,n,x2,n,x3)
axarr[0].set_title('Entradas')
axarr[0].legend(['x1','x2','x3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[0].grid()
axarr[1].plot(n,y1+y2,'--',n,y3)
axarr[1].set_title('Saídas')
axarr[1].legend(['y1+y2','y3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()

# y do Acumulador
y1 = accumulator(x1)
y2 = accumulator(x2)
y3 = accumulator(x3)

# Plots do Acumulador
fig, axarr = subplots(1, 2)
fig.suptitle("Acumulador", fontsize=16)
axarr[0].plot(n,x1,n,x2,n,x3)
axarr[0].set_title('Entradas')
axarr[0].legend(['x1','x2','x3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[0].grid()
axarr[1].plot(n,y1+y2,'--',n,y3)
axarr[1].set_title('Saídas')
axarr[1].legend(['y1+y2','y3'],loc='upper center',
                bbox_to_anchor=(0.5, -0.1),
                fancybox=True, shadow=True, ncol=5)
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
show()
```
