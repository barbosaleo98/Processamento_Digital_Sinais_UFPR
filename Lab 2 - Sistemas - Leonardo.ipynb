```python
# Inicializa os pacotes que serão utilizados
import numpy as np
import matplotlib.pyplot as plt
```

1) Implemente os seguintes sistemas: </br>

1.a ) atrasa a entrada em duas unidades

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o número de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que irá ser preenchido pela função sistA
out = np.array(np.zeros(N))

# Função sistA que desloca os valores de saída 2 unidades para a direita 
# Para i<=1 não é possível deslocar 2 unidades, pois não existe valor de
# x anterior a x[1]. Assim um if é necessário
def sistA(x,N):  
    for i in range (N):
        if i <= 1 :
            out[i] = 0
        else:
            out[i] = x[i-2]
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Atraso da entrada em duas unidades", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistA(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.b ) Média-móvel: Cada valor da saída seja a média aritmética do valor de entrada e suas 2 posições anteriores.

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o número de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que irá ser preenchido pela função sistB
out = np.array(np.zeros(N))

# Função sistB que calcula a média aritmética entre o valor 
# atual de x e 2 posições anteriores. Em i=1 e i=2 o cálculo 
# da média é realizado com a quantidade de valores disponíveis, 
# sendo que # a partir de i=3 sempre haverá pelo menos 2 
# valores anteriores disponíveis até i=N. 

def sistB(x,N):
    for i in range (N):
        if i == 1 :
            out[i] = x[i]/3
        elif i==2:
            out[i] = (x[i]+x[i-1])/3
        else:
            out[i] = (x[i]+x[i-1]+x[i-2])/3
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Média móvel", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistB(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.c ) a saída corresponde ao valor quadrático da entrada

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o número de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que irá ser preenchido pela função sistB
out = np.array(np.zeros(N))

# Função sistC cuja saída é a entrada elevada ao quadrado usando a função
# power() do pacote Numpy.
def sistC(x,N):
    for i in range (N):
        out[i] = np.power(x[i],2)
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Saída quadrática", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistC(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.d ) a saída corresponde a uma combinação linear da entrada multiplicada por 2 mais a entrada deslocada de 1 unidade multiplicada por 3

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o número de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que irá ser preenchido pela função sistB
out = np.array(np.zeros(N))

# Função sistD que realiza a combinação linear entre os vetores modificados de x.
# Para i=1 não é possível deslocar, pois não existe valor de x anterior a x[1]. 
# Assim um if é necessário
def sistD(x,N):
    for i in range (N):
        if i==1:
            out[i] = (x[i]*2)
        else:
            out[i] = (x[i]*2) + (x[i-1]*3)
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Combinação linear", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistD(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

1.e ) a saída corresponde ao dobro da entrada somada ao valor 5

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o número de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que irá ser preenchido pela função sistB
out = np.array(np.zeros(N))

# Função sistE que dobra o valor de x e soma 5
def sistE(x,N):
    for i in range (N):
        out[i] = (x[i]*2)+5
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Dobro da entrada somada ao valor 5", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistE(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```

 1.f ) Acumulador. A saída correponde a acumulação (soma) dos valores da entrada até o presente momento.

```python
# Cria o vetor x contendo os valores solicitados no enunciado
x = np.array([0, 0, 0, 1, 2, 3, 0, 0, 0])

# Determina o número de amostras de acordo com o tamanho de x
N = np.size(x)

# Cria o vetor que irá ser preenchido pela função sistB
out = np.array(np.zeros(N))

# Função sistF que soma a saída os valores da entrada anteriores. 
# Note que como não há valores anteriores a x[1], em i=1 a saída out[1] é 
# composta somente por x[1].
def sistF(x,N):
    for i in range (N):
        if i == 1:
            out[i] = x[i]
        else:
            out[i] = out[i-1] + x[i]
    return out

# Cria ambientes de figura, utilizando subplots para imprimir dois gráficos 
# em uma mesma janela.
fig, axarr = plt.subplots(2, 1)
fig.suptitle("Acumulador", fontsize=16)
axarr[0].stem(x,basefmt = " ",use_line_collection=True)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].stem(sistF(x,N),basefmt = " ",use_line_collection=True)
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show() 
```
