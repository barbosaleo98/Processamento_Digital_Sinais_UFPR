4) Calcule a frequência fundamental em cada um desses 3 casos. Plote e mostre os passos deste cálculo.

```python
def Calc_Freq(f, data):
    #Cria um ndarray com as frequencias e amplitudes correspondentes 
    # (voz humana vai de ~50 Hz a ~3400 Hz)
    sinal_teste = sp.array(list(zip(f[range(50, 3400)], data[range(50, 3400)]))) 
    
    #Acha o primeiro máximo que corresponde à uma das frequências de maior amplitude
    maxf1 = max(sinal_teste, key = itemgetter(1))[0]
    
    #Remove tudo que for anterior ao máximo + 40 Hz (assumindo que a voz humana 
    # comece em 50 Hz, só haverá múltiplos de no mínimo 50 Hz nos sinais)
    for i in range(0, len(sinal_teste)):
        if any(sinal_teste[i] == maxf1):      
            sinal_teste[range(0, i + 41)] = 0
    
    #Acha o segundo máximo que corresponde à uma harmônica
    maxf2 = max(sinal_teste, key = itemgetter(1))[0]
    
    #Subtrai uma harmônica da outra para retornar a frequência fundamental
    return int(maxf2) - int(maxf1)
    
f, Pxx_den_a = signal.welch(voz_a[:, 0], fs, 'hann', nperseg = 65535)
freq_fundamental_a = Calc_Freq(f, Pxx_den_a)

f, Pxx_den_a_agudo = signal.welch(voz_a_agudo[:, 0], fs, 'hann', nperseg = 65535)
freq_fundamental_a_agudo = Calc_Freq(f, Pxx_den_a_agudo)

f, Pxx_den_a_grave = signal.welch(voz_a_grave[:, 0], fs, 'hann', nperseg = 65535)
freq_fundamental_a_grave = Calc_Freq(f, Pxx_den_a_grave)

print('===============================================================')

print('Frequência fundamental voz natural (vogal a): {f1} Hz\n'
      .format(f1 = freq_fundamental_a))
print('Frequência fundamental voz aguda (vogal a): {f1} Hz\n'
      .format(f1 = freq_fundamental_a_agudo))
print('Frequência fundamental voz grave (vogal a): {f1} Hz\n'
      .format(f1 = freq_fundamental_a_grave))
print('===============================================================')
```
