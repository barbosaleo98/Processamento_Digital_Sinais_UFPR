Este laboratório consiste em explorar a transformada de fourier de de um sinal senoidal.

Considere um sinal x, uma cossenoide de frequencia wx, com 100 valores: n=0:99; x=cos(wx*n);

Deseja-se visualizar o espectro de x para valores de w entre 0 e 2pi, com 100 valores, nas seguintes situações. Plotar o sinal x e a magnitude da transformada de fourier (X) para os seguintes casos, quando x tem:

```python
# Inicializa os pacotes que serão utilizados
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cria o vetor n  do eixo horizontar
n = np.linspace(0,99,100)

# Cria o vetor w da frequência do sinal
w = np.linspace(0,2*np.pi,100)
```

1) exatamente 1 ciclo

```python
# Cria a variável Nciclos para armazenar a 
# quantidade de ciclos do sinal cossenoidal

Nciclos = 1

# Cria o sinal cossenoidal x e sua transformada 
# de Fourier X

x = np.cos(w*Nciclos)
X = np.fft.fft(x)


# Cria ambientes de figura, utilizando subplots para 
# imprimir dois gráficos em uma mesma janela.

fig, axarr = plt.subplots(2, 1)
fig.suptitle("Comparativo entre o sinal x e X", fontsize=16)
axarr[0].plot(n,x)
axarr[0].set_title('Entrada')
axarr[0].grid()
axarr[1].plot(n,abs(X),'orange')
axarr[1].set_title('Saída')
axarr[1].grid()
fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()

# Determina o valor máximo de X
maxElement = np.amax(abs(X))
 
print('Valor máximo de X: ', maxElement)

# Determina em qual bin está o maior valor de X
result = np.where(abs(X) == np.amax(abs(X)))

print('Bins com estes valores :', result[0])
```
